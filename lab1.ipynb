```python
attendance = datos['attendance']
count = 0
for x in attendance:
   datos['attendance'][count] = x[:-2]
   count +=1
```

```python
#eliminar el :
duracion = datos["game_duration"]
count = 0
for x in duracion:
   datos["game_duration"][count]=x[1:]
   count +=1
```

```python
#eliminar el :
venue = datos["venue"]
count = 0
for x in venue:
   datos["venue"][count]=x[1:]
   count +=1
```

```python
#modificar el time
dates = datos['date'].str.split(",")
day = []
month = []
year = []
for x in dates:
    day.append(x[0])
    month.append(x[1])
    year.append(x[2])
datos['day'] = day
datos['month'] = month
datos['year'] = year
```

```python
#eliminar la palabra Start Time
time = datos["start_time"].str.split(":",1)
count = 0
for x in datos["start_time"]:
    datos["start_time"][count] = time[count][1]
    count +=1
```
